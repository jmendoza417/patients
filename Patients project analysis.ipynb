```python
#Adding a new column, grouping by age

patients.loc[patients['Age']<=19, 'age_group'] = 'teenage'
patients.loc[patients['Age'].between(20,24), 'age_group'] = 'yadult'
patients.loc[patients['Age'].between(25,39), 'age_group'] = 'adult'
patients.loc[patients['Age']>39, 'age_group'] = 'older_adult'
```

```python
patients['age_group'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
ax = patients['age_group'].value_counts().plot(kind='bar', figsize=(6,5))
ax.set_ylabel('Billing_Amount')
```
